Tasks:

By using the flights data find all flights:

1. Had an arrival delay of two or more hours.

2. Flew to Houston (IAH or HOU)

3. Were operated by American, Delta

4. How many values are missing in `dep_time`?

5. Sort `flight` to find fastest flight.

6. Which flights travelled the shortest?

```python
import pandas as pd

flights = pd.read_csv('https://raw.githubusercontent.com/msaricaumbc/DS_data/master/nyc_flights.csv')
```

<h5>1. Had an arrival delay of two or more hours.</h5>

```python
arrival_greater_than_2 = flights['arr_delay'] >= 2
flights[arrival_greater_than_2]
```

<h5>2. Flew to Houston (IAH or HOU)</h5>

```python
flew_to_IAH_or_HOU = (flights['dest'] == 'IAH') | (flights['dest'] == 'HOU')
flights[flew_to_IAH_or_HOU]
```

<h5>3. Were operated by American, Delta<h/5>

```python
operated_by_AA_DL = (flights['carrier'] == 'AA') | (flights['carrier'] == 'DL')
flights[operated_by_AA_DL]
```

<h5>4.How many values are missing in dep_time?</h5>

```python
flights.dep_time.isnull().sum()
```

<h5> 5.Sort flight to find fastest flight.</h5>

```python
flights.sort_values(by='air_time', ascending = False)
```

<h6> 6.Which flights travelled the shortest?</h6>

```python
flights.sort_values(by='distance', ascending = True)
```
