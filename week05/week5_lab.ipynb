```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

diamonds = sns.load_dataset("diamonds")

diamonds.head()
```

#### 1.

- Plot the scatter plot of price and carat. Add clarity into the plot as hue. 

- Do you see any pattern? Is it parallel to your expectation?

```python
fig = plt.gcf()
# Change seaborn plot size
fig.set_size_inches(12, 8)
plot=sns.scatterplot(x= diamonds.carat, y = diamonds.price, hue = diamonds.clarity)
plot.set(title="price vs carat vs clarity")
```

<h5>Observations</h5>
<ol>
    <li>Higher carats have lowest clarity.</li>
    <li>Majority of the diamonds are of S12 calrity</li>
    <li>Diamonds with carat value greater than 2.5 are very few</li>
    <li>More than 95% of the diamonds are less than 2.5 carats</li>
</ol>

#### 2.

- Consider `clarity` and `color` is there any pattern in their heatmap?

```python
plot = sns.displot(diamonds, x="clarity", y="color", cbar = True, aspect=1.5)
plot.ax.set_title('Color and Clarity Heat map')
```

<h5>Pattern Observed</h5>
<ul>
    <li>Most of the diamonds are between VS1 clarity and i1</li>
</ul>

#### 3.

- Use `carat` variable as as the size of the points in scatter plot in which x-axis is the `depth` and `price` on the y-axis.

```python
fig = plt.gcf()
# Change seaborn plot size
fig.set_size_inches(12, 8)
plot=sns.scatterplot(data = diamonds, x="depth", y="price", size='carat')
plot.set(title="Depth vs Carat")
```

<h5>Observation</h5>
Most of the diamonds are between 1 and 3 carats
