```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

diamonds = sns.load_dataset("diamonds")

diamonds.head()
```

#### 1.

- Plot the scatter plot of price and carat. Add clarity into the plot as hue. 

- Do you see any pattern? Is it parallel to your expectation?

```python
# sns.scatterplot(x= diamonds.price, y = diamonds.carat, hue = diamonds.clarity, alpha = 0.1)
sns.scatterplot(data = diamonds, x="carat", y="price", hue='clarity')
plt.show()
```

<h5>Observations</h5>
<ol>
    <li>Higher carats have lowest clarity.</li>
    <li>Majority of the diamonds are of S12 calrity</li>
    <li>Diamonds with carat value greater than 2.5 are very few</li>
    <li>More than 95% of the diamonds are less than 2.5 carats</li>
</ol>
