<h4>imports</h4>

```python
import pandas as pd
import pandas as pd
from bs4 import BeautifulSoup
import requests
import re
import numpy as np
```

```python
country_list_df = pd.read_csv('countrydata.csv')
country_list_df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
response = requests.get('https://en.wikipedia.org/wiki/List_of_countries_by_past_and_projected_GDP_(nominal)')
response.ok
```

```python
html_doc = response.text
soup = BeautifulSoup(html_doc, 'html.parser')
```

```python
t = soup.find_all('table')[5]
t=t.find_all('tr')
```

```python
data=[[y.text.lstrip('\xa0').rstrip().replace(',', '') for y in x.find_all('td')] for x in t]
```

```python
df2=pd.DataFrame(data, columns=['CountryName', "Year 2020", "Year 2021", "Year 2022", "Year 2023", "Year 2024", "Year 2025", "Year 2026", "Year 2027"])
```

```python
df2 = df2.iloc[1: , :]
df2.drop(df2.columns[[1, 2]], axis=1, inplace=True)
country_list_df=country_list_df.sort_values(by='CountryName')
```

```python
combined_df=pd.merge(country_list_df, df2, left_on="CountryName", right_on="CountryName")
```

```python
combined_df[["Year 2022", "Year 2023", "Year 2024", "Year 2025", "Year 2026", "Year 2027"]]=combined_df[["Year 2022", "Year 2023", "Year 2024", "Year 2025", "Year 2026", "Year 2027"]].apply(pd.to_numeric)
```

```python
def annualGrowthRate(a, b, c, d, e, f):
    return ((((b-a)/a)*100)+(((c-b)/b)*100)+(((d-c)/c)*100)+(((e-d)/d)*100)+(((f-e)/e)*100))/5
```

```python
combined_df['average_rate'] = combined_df.apply(lambda row : annualGrowthRate(row['Year 2022'],
                     row['Year 2023'], row['Year 2024'], row['Year 2025'], row['Year 2026'], row['Year 2027']), axis = 1)
```
