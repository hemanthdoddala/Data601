Please complete the functions below. These functions are for to simulate a fair and a biased die. 

The biased die has probabilities {0.20, 0.10, 0.15, 0.15, 0.15, 0.25}.

Create 2 visualizations that shows outcomes of multiple rolls of a fair die and biased die.

Repeat the process programatically and take the mean to create a new list. This list will be used to compare means of biased & unbiased dice.


````python
import random
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

trial_count = 500

def fair_die_simulation(trials):
    """
    returns an array of randomly numbers between 1 and 6. 
    """
    
    options = [1,2,3,4,5,6]
    simulations = [random.choice(options) for _ in range(trials)]
    return simulations
````

````python
def biased_die_simuations(trials, weights=[0.20, 0.10, 0.15, 0.15, 0.15, 0.25]):
    """
    returns an array of randomly numbers between 1 and 6 with the probabilty of assosiated weight. 
    for instance: for weights: [0.25, 0.15, 0.15, 0.15, 0.15, 0.15]
    1 has 0.25 change to be selected, 2 has 0.15 chance and so on...
    """
    
    options = [1,2,3,4,5,6]
    simulations = random.choices(options, weights=weights,k=trials)
    return simulations
````

````python
biased = biased_die_simuations(trial_count)
fair = fair_die_simulation(trial_count)
fair_df= pd.DataFrame([{"1's count":fair.count(1),"2 count":fair.count(2), "3 count":fair.count(3), "4 count":fair.count(4), "5 count":fair.count(5), "6 count":fair.count(6)}])
biased_df= pd.DataFrame([{"1's count":biased.count(1),"2 count":biased.count(2), "3 count":biased.count(3), "4 count":biased.count(4), "5 count":biased.count(5), "6 count":biased.count(6)}])
fair_df
biased_mean_list = [np.mean(biased_die_simuations(trial_count)) for _ in range(trial_count)]
unbiased_mean_list = [np.mean(fair_die_simulation(trial_count)) for _ in range(trial_count)]
````

````python
biased_df.plot.bar()
plt.title("Biased Dice Bar Graph")
plt.xticks([])
plt.ylabel("Count")
plt.show()
````

````python
fair_df.plot.bar()
plt.title("Fair Dice Bar Graph")
plt.xticks([])
plt.ylabel("Count")
plt.show()
````

````python
plt.figure(figsize=(15, 7))
plt.subplot(2, 2, 1)
plt.hist(biased_mean_list)
plt.title("Biased Dice Mean List")
plt.xlabel("mean")
plt.ylabel("frequency")
plt.subplot(2, 2, 2)
plt.hist(unbiased_mean_list, color='orange')
plt.title("Fair Dice Mean List")
plt.xlabel("mean")
plt.ylabel("frequency")
plt.subplot(2, 2, (3, 4))
plt.hist(biased_mean_list, color='green', label="biased")
plt.hist(unbiased_mean_list, color='red', label='unbiased')
plt.title("Biased vs Fair Dice")
plt.xlabel("mean")
plt.ylabel("frequency")
plt.legend()
plt.show()
````
