# Lab

Using the flights dataframe answer the following questions

1. When do the first and last flights leave each day?

2. When there is a missing value in `dep_time` then this is an indication of canceled flight. Find the number of cancelled flights for each (`carrier`, `origin`) combination.

3. Which carrier has the worst `dep_delay`s? 

4. Which plane (`tailnum`) has the worst on-time record?

5. For each plane, count the number of flights before the first delay of greater than 1 hour.

```python
import pandas as pd
url = 'https://raw.githubusercontent.com/msaricaumbc/DS_data/master/nyc_flights.csv'

flights = pd.read_csv(url)
```

```python
flights.head(10)
```

<h5>1.When do the first and last flights leave each day?</h5>

```python
multiple_grouped = flights.groupby(['year', 'month', 'day']).agg(first_flight = ('dep_time', 'min'), last_flight = ('dep_time', 'max'))
multiple_grouped.head(10)
```

```python
multiple_grouped.tail(10)
```

<h5>2. When there is a missing value in dep_time then this is an indication of canceled flight. Find the number of cancelled flights for each (carrier, origin) combination.</h5>

```python
flights.groupby(['carrier', 'origin'])['dep_time'].apply(lambda x: x.isnull().sum())
```

<h5>3. Which carrier has the worst `dep_delay`s?</h5> 

```python
grp_data=flights.groupby(['carrier']).agg(average_delay=('dep_delay', 'mean'))
print(grp_data)
```

```python
print(grp_data.loc[grp_data['average_delay'].idxmax()])
# or 
grp_data.sort_values(by='average_delay', ascending=False).head(1)
```

<h5>4. Which plane (`tailnum`) has the worst on-time record?</h5>

```python
flights.groupby(['tailnum']).agg(average_delay=('arr_delay', 'mean')).sort_values(by='average_delay', ascending=False).head(1)
```

<h5>5. For each plane, count the number of flights before the first delay of greater than 1 hour.</h5>

```python
flights[flights['dep_delay'] > 60].groupby('tailnum')['dep_delay'].count()
```
